# Calculate salary after taxes

Salary after taxes = Salary * (1 - tax rate)

Monthly salary = Salary after taxes / 12

```python
salary = float(input('Initial Salary?'))
tax_rate = float(input('Tax rate?'))

salary_after_taxes = salary * (1 - tax_rate)
print('Salary after taxed: ' + str(round(salary_after_taxes,2)))
monthly_salary = salary_after_taxes / 12
print('Monthly salary: ' + str(round(monthly_salary,2)))
```

```python
# Monthly Expenses and Savings

monthly_rent = 1200
daily_food_budget = 30
monthly_food_budget = 30*30
monthly_entertainment_budget = 200
monthly_unforeseen_expenses = 250

monthly_expenses = monthly_rent + monthly_food_budget + monthly_entertainment_budget + monthly_unforeseen_expenses
print("Monthly expenses: " + str(round(monthly_expenses, 2)))

monthly_savings = monthly_salary - monthly_expenses
print("Monthly savings: " + str(round(monthly_savings, 2)))
```

## Forecast Salary Growth and Cost of Living

There's inflation so there is annual increase in your salary.

Convert annual salary growth to monthly salary growth

### r_monthly = (1 + r_annual)^(1/12)-1

```python
# Initial salary after divide 12
monthly_takehome_salary = float(input('How much you made?'))

# Monthly forecast
year = int(input('How many years?'))
forecast_months = 12 * year

# Annual salary growth rate
annual_salary_growth = float(input('Estimate rate?'))

# Monthly salary growth rate
monthly_salary_growth = (1 + annual_salary_growth)**(1/12) - 1

# Forecast cumulative growth of salary
cumulative_salary_growth_forecast = np.cumprod(np.repeat(1+monthly_salary_growth,forecast_months))

# Calculate the actual salary forecast
salary_forecast = monthly_takehome_salary * cumulative_salary_growth_forecast

# Plot forecasted salary
plt.plot(salary_forecast)
plt.xlabel('Periodt. (In months)')
plt.ylabel('Monthly Salary')
plt.show()
```

## Forecast Growing Expenses Due to Inflation

Due to Inflation so the expenses will increase too

### monthly Inflation = (1 + annual_inflation)^(1/12)-1

```python
# Initial inflation
annual_inflation = float(input('Estimate inflation? in %'))

# Monthly Expenses
monthly_expenses = float(input('Monthly expenses?'))

# Monthly forecast
year = int(input('How many years?'))
forecast_months = 12 * year

# Calculate equivalent monthly inflation rate
monthly_inflation = (1 + annual_inflation)**(1/12)-1

# Forecast cumulative inflation
cumulative_inflation_forecast = np.cumprod(np.repeat(1 + monthly_inflation,forecast_months))

# Calculate forecasted expenses
expense_forecast = monthly_expenses * cumulative_inflation_forecast 

# Plot forecasted salary
plt.plot(expense_forecast)
plt.xlabel('Periodt. (In months)')
plt.ylabel('Monthly expenses')
plt.show()
```
