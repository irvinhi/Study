# Pyplot

```python
import matplotlib.pyplot as plt
import pandas as pd 
data = pd.read_csv('stock_data.csv')
data2 = pd.read_csv('stock_data2.csv')
print(data.head())
print(data2.head())
```

# Basic Line Plot

```python
plt.plot('Day', 'Price', color='red', linestyle= '--', data=data) #Basic

#Label
plt.xlabel('Days')
plt.ylabel('Prices, $')

#title
plt.title('Company Stock Prices Over Time')
plt.show()
```

# Multiple lines on same plot

```python
plt.plot('day','company1',color='red',data=data2)
plt.plot('day','company2',color='green',data=data2)
plt.xlabel('Days')
plt.ylabel('Prices')
plt.title('Stock Prices Over Time')
plt.show()
```

# Scatter Plot

```python
plt.scatter('Day','Price',color='red',s=0.1,data=data)
#s = size of marker
```

# Histogram plt.hist()

To tell distribution of the data

Advantages: 
<li> To know shape and distribution of the data </li>
<li> To know lot of variability in data </li>
<li> Identify outliers (Data that have abnormal distances)



```python
#bins = set total range of data set
plt.hist('Price', bins=100, data=data)
plt.show()
```

```python
# Comparing two histograms

#alpha = set transparency
plt.hist('company1', bins=100, data=data2,alpha=0.5, label='Stock A') 
plt.hist('company2', bins=100, data=data2,alpha=0.5, label='Stock B')

plt.legend()
plt.show()
```
